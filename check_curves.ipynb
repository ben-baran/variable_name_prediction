```python
import glob
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = 20, 10
```

```python
plots = []
names = [] 
for fname in glob.glob('manual/*/validation_losses.npy'):
    plots.append(np.load(fname))
    names.append(fname)
```

```python
for plot in plots:
    plt.plot(plot[:100])
plt.show()

smallest_loss = float('inf')
smallest_loss_name = ''
index = 0
for plot_i, plot in enumerate(plots):
    print(names[plot_i], 'has smallest loss', plot.min(), 'at', plot.argmin())
    with open('/'.join(names[plot_i].split('/')[:-1]) + '/run_options.json', 'r') as option_file:
        print("Run options:\n", option_file.read())
    if plot.min() < smallest_loss:
        smallest_loss = plot.min()
        index = plot.argmin()
        smallest_loss_name = names[plot_i]
print("smallest loss: %f from %s, iteration %d" % (smallest_loss, smallest_loss_name, index))
```
